```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import igraph as ig
```

```python
def find_head(visited, dim):
    for i in range(dim):
        if i not in visited:
            return i
    
    return -1

def find_connection_probability(graph, sample):
    connections = 0
    for i in range(len(sample)):
        for k in range(i+1, len(sample)):
            if(len(graph.get_shortest_path(sample[i].index, sample[k].index)) != 0):
                connections += 1
                
    return connections / math.comb(len(sample), 2)

def percolation(graph, p_vals, sample_size):
    connect_probs = np.zeros_like(p_vals)

    for i in range(len(p_vals)):
        graph_copy = graph.copy()

        edges = list(graph_copy.es)
        to_delete = []
        for edge in edges:
            if (random.random() < p_vals[i]):
                to_delete.append(edge.index)

        graph_copy.delete_edges(to_delete)

        sampled_nodes = random.sample(list(graph_copy.vs), sample_size)

        connect_probs[i] = find_connection_probability(graph_copy, sampled_nodes)

    return connect_probs
```

```python
p_vals = np.arange(0,1.001, 0.05)

graph = ig.Graph.Barabasi(500, 2)
connect_probs5h = percolation(graph, p_vals, 500)

graph = ig.Graph.Barabasi(1000, 2)
connect_probs1k = percolation(graph, p_vals, 1000)

graph = ig.Graph.Barabasi(5000, 2)
connect_probs5k = percolation(graph, p_vals, 1000)

#graph = ig.Graph.Barabasi(10000, 2)
#connect_probs10k = percolation(graph, p_vals, 1000)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(p_vals, connect_probs5h, color='green', label='N = 500')
ax.plot(p_vals, connect_probs1k, color='blue', label='N = 1000')
ax.plot(p_vals, connect_probs5k, color='red', label='N = 5000')
#ax.plot(p_vals, connect_probs10k, color='purple', label='N = 10000')
#ax.plot(p_vals, connect_probs, color='black', label='N = 50000')

ax.set_title("BA Model (m = 2), Sample of 1000")
ax.set_xlabel("Severing Probability")
ax.set_ylabel("Connection Probability for Random Nodes")
ax.set_yscale('log')

ax.legend()
```
